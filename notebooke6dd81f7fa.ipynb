```python
#Veri işleme için
import pandas as pd 
import numpy as np
#veriyi görselleştirmek için 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv("/kaggle/input/2023-world-population-by-country/countries-table.csv")
```

```python
#En Kalabalık 10  Ülke 
top10=df.sort_values(by="pop2023",ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x="pop2023",y="country",data=top10,palette="viridis")

plt.title("2023 Yılında En Kalabalık 10 Ülke ")
plt.xlabel("Nüfus(miyar)")
plt.ylabel("Ülke")
plt.show()
```

```python
# En hızlı büyüyen top 10 ülke
top10_growth = df.sort_values("growthRate", ascending=False).head(10)

# Grafik boyutu
plt.figure(figsize=(10,6))

# Lollipop chart
plt.hlines(y=top10_growth['country'], xmin=0, xmax=top10_growth['growthRate'], color='skyblue', linewidth=3)
plt.plot(top10_growth['growthRate'], top10_growth['country'], "o", color='orange', markersize=8)

# Başlık ve etiketler
plt.xlabel("Büyüme Oranı (%)")
plt.ylabel("Ülke")
plt.title("2023 Yılında En Hızlı Büyüyen 10 Ülke")

plt.show()
```

```python
#seaborn un gelecek sürümlerinde yaşanacak sorunu gösteren hatay ıgizledim
import warnings 
warnings.filterwarnings("ignore")


fastest_growth = df.loc[[df['growthRate'].idxmax()]]  # çift köşeli parantez -> DataFrame

plt.figure(figsize=(6,4))
sns.barplot(x="growthRate", y="country", data=fastest_growth, palette="rocket")
plt.xlabel("Büyüme Oranı (%)")
plt.ylabel("Ülke")
plt.title("En Hızlı Büyüyen Ülke (2023)")
plt.xlim(0, fastest_growth['growthRate'].values[0]*1.2)
plt.show()
```
